```python
import pandas as pd
import glob
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

```python
occupancy_model_zones = [
    'ZonePeopleOccupantCount',
    'SiteDayTypeIndex',
    'SiteHorizontalInfraredRadiationRatePerArea',
    'SiteOutdoorAirDewpointTemperature',
    'SiteOutdoorAirDrybulbTemperature',
    'SiteOutdoorAirRelativeHumidity',
    'SiteOutdoorAirWetbulbTemperature',
]
```

```python
hvac_model_zones = [
    'ZonePeopleOccupantCount',
    'SiteDayTypeIndex',
    'SiteHorizontalInfraredRadiationRatePerArea',
    'SiteOutdoorAirDewpointTemperature',
    'SiteOutdoorAirDrybulbTemperature',
    'SiteOutdoorAirRelativeHumidity',
    'SiteOutdoorAirWetbulbTemperature',
    'ElectricityHVAC'
]
```

```python
file_dict = {}

for f in glob.glob("2004/run_*/*"):
    run_parts = f.split('/')
    run_name = run_parts[1]
    zone = run_parts[2] 
    print(run_name, zone)
    if zone in occupancy_model_zones:
    
        if run_name not in file_dict:
            file_dict[run_name] = {}
        
        for f1 in glob.glob(f'{f}/*'):
            if 'block1_items' in f1:
                df = pd.DataFrame()
                temp = pd.read_csv(f1)
                column_names = temp.values.flatten()
                df[column_names] = None
            
            if 'block1_values' in f1:
                temp = pd.read_csv(f1, names=column_names)
                df[column_names] = temp.values
                file_dict[run_name][zone] = df
                print(f1, df.shape)
                
for run in file_dict.keys():
    final_df = pd.DataFrame()
    for zone in file_dict[run].keys():
        # display(file_dict[run][zone].head())
        final_df = pd.concat([final_df, file_dict[run][zone]], axis=1)
        print(final_df.shape)
    
    # final_df.to_csv(f'{run}.csv', index=False)
    # print(f'{run}.csv saved')
    final_df.to_csv(f'occupancy_model_zones.csv', index=False)
    print(f'occupancy_model_zones.csv saved')
```
